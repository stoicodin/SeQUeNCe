```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("distance_bb84_link_trial.csv")  #Change 'file.csv' to your csv file name

# Create a line plot 
plt.plot(data['Distance'], data['Error_rate'])

# Add data points onto the line plot
plt.scatter(data['Distance'], data['Error_rate'], color='red', marker='.', s=20)

# Add labels for x and y axes
plt.ylabel("Error rate")
plt.xlabel("Distance(m)")
plt.grid()
# Display the plot
plt.show()
```

```python


# Assume the data has columns named 'A' and 'B'
plt.plot(data['Distance'], data['Throughput'])
plt.ylabel("Throughput")
plt.xlabel("Distance(m)")
plt.scatter(data['Distance'], data['Throughput'], color='red', marker='.', s=20)
plt.grid()
plt.show()
```

```python
# Assume the data has columns named 'A' and 'B'
plt.plot(data['Distance'], data['Latency'])
plt.ylabel("Latency")
plt.xlabel("Distance(m)")
plt.scatter(data['Distance'], data['Latency'], color='red', marker='.', s=20)
plt.grid()
plt.show()
```

```python
import numpy as np  # Needed for log function

# Load the data
data = pd.read_csv("distance_bb84.csv")  #Change 'file.csv' to your csv file name

# Compute the natural logarithm of the Throughput data
log_throughput = np.log(data['Throughput'])

# Create a line plot 
plt.plot(data['Distance'], log_throughput)

# Add data points onto the line plot
plt.scatter(data['Distance'], log_throughput, color='red', marker='.', s=20)

# Add labels for x and y axes
plt.ylabel("Log of Throughput")
plt.xlabel("Distance(m)")
plt.grid()
# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("distance_bb84_200m.csv")  

fig, axs = plt.subplots(1, 3, figsize=(15,5))

# Create a line plot 
axs[0].plot(data['Distance']/1000, data['Error_rate']*100)
axs[0].scatter(data['Distance']/1000, data['Error_rate']*100, color='red', marker='.', s=20)
axs[0].set_xlabel('Distance(km)')
axs[0].set_ylabel('Error rate(%)')
axs[0].grid()

axs[1].plot(data['Distance']/1000, data['Throughput']/(10**6))
axs[1].scatter(data['Distance']/1000, data['Throughput']/(10**6), color='red', marker='.', s=20)
axs[1].set_xlabel('Distance(km)')
axs[1].set_ylabel('Throughput(MHz)')
axs[1].grid()

axs[2].plot(data['Distance']/1000, data['Latency'])
axs[2].scatter(data['Distance']/1000, data['Latency'], color='red', marker='.', s=20)
axs[2].set_xlabel('Distance(km)')
axs[2].set_ylabel('Latency(s)')
axs[2].grid()

plt.tight_layout() # This helps to fix issues with labels being cut-off
plt.show()
```
